```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df = pd.read_csv("/content/drive/MyDrive/MENTAL-LOAD-DATASET/final.csv")
# drop first column
df = df.drop(df.columns[0], axis = 1)

df
```

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
def add_noise(signal):
    noise=np.random.normal(0,0.5,3840)
    return (signal+noise)
```

```python
noise_df = X.apply(add_noise, axis = 1)
```

```python
# add labels
noise_df = noise_df.assign(label = y)
noise_df
```

```python
# combine noise dataset with original dataset
final_df = df.append(noise_df)
final_df
```

```python
# convert df to csv
final_df.to_csv("/content/drive/MyDrive/MENTAL-LOAD-DATASET/data-with-noise.csv")
```
